```python
import pathlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import PIL
import shutil
```

```python
path_to_pets = pathlib.Path("downloads", "pets")
```

```python
df = pd.DataFrame(path_to_pets.glob("*"), columns=["path"])
```

```python
df
```

```python
fig, axs = plt.subplots(5, 3, figsize=(8, 6))

df_sample = df.sample(np.prod(axs.shape))

for i, ax in enumerate(axs.ravel()):
    img_path = df_sample.iloc[i].path
    ax.imshow(PIL.Image.open(img_path))
    ax.axis(False)
    ax.set_title(img_path.name, fontdict={'fontsize': 10})
plt.tight_layout()
plt.show()
```
